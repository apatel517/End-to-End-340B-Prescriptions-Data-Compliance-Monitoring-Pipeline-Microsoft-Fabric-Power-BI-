```python
# Welcome to your new notebook
# Type here in the cell editor to add code!
from pyspark.sql.functions import input_file_name, to_date, col ,current_timestamp
from pyspark.sql.types import DoubleType, IntegerType

# 1 - Location Data 

location_path = "fakepath/*.csv"
df_location = spark.read.option("header" , True).csv(location_path)
df_location = (
    df_location
    .withColumn("source_file", input_file_name())
    .withColumn("Time_Stamps", current_timestamp())
)
df_location.show(n=5, truncate=False)

df_location.write.format("delta").mode("append").option("mergeSchema", "true").saveAsTable("bronze_Data_Location")
```
